# Import

```python
import pandas as pd
import numpy as np
```

# Read CSV

```python
dataset = pd.read_csv('spotify_data clean.csv')
```

# Check the data from the first two columns

```python
dataset.head(2)
```

# Checking Null values

```python
dataset.isnull().sum()
```

# Removing Null Values

```python
dataset = dataset.dropna()
```

# After Removing again checking Null values

```python
dataset.isnull().sum()
```

# Checking Data Type

```python
dataset.dtypes
```

# Checking Duplication

```python
dataset.duplicated().sum()
```

# If there is Duplicate then Removing

```python
dataset = dataset.drop_duplicates(subset=['track_id'])
```

# Changing DataType of the 'album_release_date' from Object to Date

```python
dataset['album_release_date'] = pd.to_datetime(dataset['album_release_date'])
```

# After changing the data type, check the data again

```python
dataset.dtypes
```

# Changing the 'explicit' Data Type from Boolean to Integer (0 and 1)

```python
dataset['explicit'] = dataset['explicit'].astype(int)
```

```python
dataset.dtypes
```

```python
# Make CSV File
```

```python
import os
os.makedirs("Clean_Datasets", exist_ok=True)
dataset.to_csv("Clean_Datasets/spotify_cleaned_data.csv", index=False)
```
